```python
#Program sederhana Tahun Kabisat
tahun = int(input("Masukkan tahun: "))

if (tahun % 400 == 0 ):
    print(f"{tahun} adalah tahun kabisat.")
else:
    if (tahun % 100 == 0):
        print(f"{tahun} bukan tahun kabisat.")
    else:
        if (tahun % 4 == 0):
            print(f"{tahun} adalah tahun kabisat.")
        else:
            print(f"{tahun} bukan tahun kabisat.")
```

```python
#Program sederhana angka terbesar
no_1 = int(input("Masukkan angka pertama: "))
no_2 = int(input("Masukkan angka kedua: "))
no_3 = int(input("Masukkan angka ketiga: "))

print(f"Angka pertama adalah {no_1}")
print(f"Angka kedua adalah {no_2}")
print(f"Angka ketiga adalah {no_3}")

if (no_1 >= no_2) and (no_1 >= no_3):
    print(f"Jadi, {no_1} adalah angka terbesar.")
else:
    if (no_2 >= no_1) and (no_2 >= no_3):
        print(f"Jadi, {no_2} adalah angka terbesar.")
    else:
        print(f"Jadi, {no_3} adalah angka terbesar.")
```

```python
#Program sederhana menghitung jumlah bayar konsumen
harga = int(input("Masukkan jumlah harga: "))
dis_tam_mem = int(harga * 2 / 100)

if harga >= 200000:
    diskon = int(harga * 4 / 100)
    print(f"Member harus membayar sebesar {harga - diskon - dis_tam_mem}")
    print(f"Non-Member harus membayar sebesar {harga - diskon}")
else:
    if harga >= 100000:
        diskon = int(harga * 3 / 100)
        print(f"Member harus membayar sebesar {harga - diskon - dis_tam_mem}")
        print(f"Non-Member harus membayar sebesar {harga - diskon}")
    else:
        print(f"Member harus membayar sebesar {harga - dis_tam_mem}")
        print(f"Non-Member harus membayar sebesar {harga}")
```

```python
#Program membuat user baru
username = input("Masukkan username: ")
email = input("Masukkan email: ")
password = input("Masukkan password: ")

if (username == ""):
    print("Username tidak boleh kosong.")
else:
    if (email == ""):
        print("Email tidak boleh kosong.")
    elif "@" not in email:
        print("Email tidak valid.")
    elif (password == ""):
        print("Password tidak boleh kosong.")
    elif len(password) < 8:
        print("Password harus memiliki minimal 8 karakter.")
    else:
        print("User berhasil dibuat.")
```
