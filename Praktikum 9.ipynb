```python
#Program untuk mengelola data
data = {
        "Nama": "Dzakii Luqman Faid",
        "Kelas": "1",
        "Tanggal Lahir": "27-05-2006",
        "No HP": "08080808",
        "Alamat": "Jl Nanas Pajangan",
        "Hobi": ["Hobi 1", "Hobi 2", "Hobi 3"]
    }

for key, value in data.items():
  if isinstance(value, list):
    print(f"{key}:")
    for index, hobi in enumerate(value, start=1):
      print(f" {index}. {hobi}")
  else:
    print(f"{key}: {value}")
```
