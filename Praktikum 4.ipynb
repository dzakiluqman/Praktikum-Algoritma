```python
nilai = input("Masukkan nilai Anda: ")
nilai = int(nilai)

if nilai >= 90:
  print("A")
elif nilai >= 80:
  print("B+")
elif nilai >= 70:
  print("B")
elif nilai >= 60:
  print("C+")
elif nilai >= 50:
  print("C")
elif nilai >= 40:
  print("D")
else:
  print("E")
```

```python
nilai = input("Masukkan nilai Anda: ")
nilai = float(nilai)

if 80 <= nilai <= 100:
  print("A")
elif 77.50 <= nilai <= 79.99:
  print("A-")
elif 75.00 <= nilai <= 77.49:
  print("AB")
elif 72.50 <= nilai <= 74.99:
  print("B+")
elif 70.00 <= nilai <= 72.49:
  print("B")
elif 67.50 <= nilai <= 69.99:
  print("B-")
elif 65.00 <= nilai <= 67.49:
  print("BC")
elif 62.50 <= nilai <= 64.99:
  print("C+")
elif 60.00 <= nilai <= 62.49:
  print("C")
elif 55.00 <= nilai <= 59.99:
  print("C-")
elif 50.00 <= nilai <= 54.99:
  print("CD")
elif 45.00 <= nilai <= 49.99:
  print("D+")
elif 40.00 <= nilai <= 44.99:
  print("D")
else:
  print("E")
```
