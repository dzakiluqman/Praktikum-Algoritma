```python
# Weekly Exercise 1
class AlatTulis:
    def __init__(self):
        self.nama = ""
        self.stok = 0
        self.harga_satuan = 0
        self.harga_total = 0

    def set_data(self, nama, stok, harga_satuan):
        self.nama = nama
        self.stok = stok
        self.harga_satuan = harga_satuan
        self.harga_total = stok * harga_satuan

    def get_data(self):
        return {
            "Nama": self.nama,
            "Stok": self.stok,
            "Harga Satuan": self.harga_satuan,
            "Harga Total": self.harga_total
        }


class Transaksi:
    def __init__(self):
        self.total_harga = 0

    def set_total_harga(self, alat_tulis_list):
        self.total_harga = sum([alat.harga_total for alat in alat_tulis_list])

    def get_total_harga(self):
        return self.total_harga

if __name__ == "__main__":
    bolpoint = AlatTulis()
    pensil = AlatTulis()
    penghapus = AlatTulis()

    bolpoint.set_data("Bolpoint", 10, 2000)
    pensil.set_data("Pensil", 10, 1000)
    penghapus.set_data("Penghapus", 10, 500)

    for alat in [bolpoint, pensil, penghapus]:
        data = alat.get_data()
        print(f"Nama: {data['Nama']}, Stok: {data['Stok']}, Harga Satuan: Rp.{data['Harga Satuan']}, Harga Total: Rp.{data['Harga Total']}")

    transaksi = Transaksi()
    transaksi.set_total_harga([bolpoint, pensil, penghapus])

    print(f"\nTotal Harga Semua Barang: Rp.{transaksi.get_total_harga()}")
```
