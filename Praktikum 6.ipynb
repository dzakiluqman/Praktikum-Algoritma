```python
# For loop example

dates = [1990,1980,1970]
N = len(dates)

for i in range(N):
    print(dates[i])
```

```python
# While Loop Example

dates = [1990, 1980, 1970, 1960]

i = 0
year = 0

while(year != 1970):
    year = dates[i]
    i = i + 1
    print(year)

print("Butuh perulangan sebanyak ", i)
```

```python
baris = int(input("Masukkan jumlah baris: "))
spasi = baris - 1

for i in range(baris):
  print(" " * spasi + "* " * (i+1))
  spasi -= 1
```

```python
x = int(input("Masukkan jumlah baris: "))

def pascal(n, k):
  if k in (0, n):
    return 1
  return pascal(n - 1, k - 1) + pascal(n - 1, k)

for row in range(x):
  print(' ' * (x - row - 1), end='')
  for k in range(row + 1):
    print(pascal(row, k), end=' ')
  print()
```

```python
def pascal(n, k):
  if k in (0, n):
    return 1
  return pascal(n - 1, k - 1) + pascal(n - 1, k)

def jumlahSatuBaris(x):
    jumlah = 0
    for k in range(x):
        jumlah += pascal(x-1, k)
    return jumlah

x = int(input("Masukkan baris: "))
print(jumlahSatuBaris(x))
```
